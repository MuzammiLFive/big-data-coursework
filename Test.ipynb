```python
import pandas as pd
```

```python
# Housing sales prices of properties in Manhattan from August 2012 to 2013.
dataset = pd.read_csv("Manhattan12.csv")
```

### Data Cleaning

```python
df = dataset[4:-1]
df.columns = dataset.iloc[3]
df.head()
```

```python
df.columns
```

```python
# Removing ' ', '\n' 
updated_column_names = []
for column in df.columns:
    column = column.replace(" ", "_").replace("\n","").replace("-","")
    updated_column_names.append(column)
df.columns = updated_column_names
```
